```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

##Q.1
M = float(input("Enter the value of M: "))
x = np.linspace(-10, 10, 400)

y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, 'r--', label=r'$y = Mx^2$')
plt.plot(x, y2, 'b-', label=r'$y = M \sin(x)$')

plt.legend()
plt.grid(True)
plt.title(f"Plots for M = {M}")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
##Q.3

import numpy as np
import matplotlib.pyplot as plt

roll_number = 102306092
np.random.seed(roll_number)


data = np.random.randn(50)


cumulative_sum = np.cumsum(data)


fig, axs = plt.subplots(2, 2, figsize=(10, 8))


axs[0, 0].plot(cumulative_sum, 'g-', label="Cumulative Sum")
axs[0, 0].set_title("Cumulative Sum Plot")
axs[0, 0].legend()
axs[0, 0].grid(True)


axs[0, 1].scatter(range(50), data, c="blue", alpha=0.6)
axs[0, 1].set_title("Random Noise Scatter Plot")
axs[0, 1].grid(True)


axs[1, 0].axis('off')
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
## Q.4


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)


plt.figure(figsize=(8,5))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', color='r')
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()


plt.figure(figsize=(10,6))
for col in df.columns[1:6]:
    sns.lineplot(x=df['month_number'], y=df[col], label=col)

plt.title("Sales Data of Different Products Over Time")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.legend()
plt.grid(True)
plt.show()


df.plot(kind='bar', figsize=(10,6), width=0.8)
plt.title("Bar Chart for Company Sales Data")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
