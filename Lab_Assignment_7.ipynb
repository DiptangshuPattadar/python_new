```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)  # Change this to your roll number

# Generate Sales Data
sales_data = np.random.randint(1000, 5000, (12, 4))

# Convert to DataFrame
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df = pd.DataFrame(sales_data, columns=categories, index=months)
```

```python
# Display first 5 rows and summary statistics
print(df.head())
print(df.describe())

# Total sales per category and per month
df['Total Sales'] = df.sum(axis=1)
total_sales_per_category = df[categories].sum()

df['Growth Rate'] = df['Total Sales'].pct_change() * 100

# Apply Discount
roll_number = 42  # Replace with your roll number
if roll_number % 2 == 0:
    df['Electronics'] *= 0.9  # 10% discount for even roll number
else:
    df['Clothing'] *= 0.85  # 15% discount for odd roll number
```

```python
df[categories].plot(figsize=(10, 6), marker='o')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()

# Box plots to show sales distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution by Category')
plt.show()
```

```python
# Q2: Array operations
array = np.array([[1, -2, 3], [-4, 5, -6]])
print("Absolute values:\n", np.abs(array))
```

```python
# Percentiles
percentiles = np.percentile(array.flatten(), [25, 50, 75])
print("Percentiles (flattened):", percentiles)
```

```python
# Mean, Median, Std Dev
print("Mean:", np.mean(array))
print("Median:", np.median(array))
print("Std Dev:", np.std(array))
```

```python
# Q4: Swap elements in a list using a temporary variable
def swap_list_elements(lst, i, j):
    temp = lst[i]
    lst[i] = lst[j]
    lst[j] = temp
    return lst

my_list = [1, 2, 3, 4, 5]
print("Swapped list:", swap_list_elements(my_list, 1, 3))
```

```python
# Q5: Swap elements in a set by converting to a list
def swap_set_elements(s, i, j):
    lst = list(s)
    lst[i], lst[j] = lst[j], lst[i]
    return set(lst)

my_set = {10, 20, 30, 40}
print("Swapped set:", swap_set_elements(my_set, 0, 2))
```
