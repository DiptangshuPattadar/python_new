Q.3 You are given a weekly sales dataset and need to perform various operations
using NumPy broadcasting.
1.  Generate your unique sales dataset:
* Take the sum of the ASCII values of the initials of your first and last
name. Call this value X. (If your initials are A B → ASCII sum = 65
+ 66 = 131 → sales = [131, 181, 231, 281, 331].)
* Create a NumPy array sales with values [X, X+50, X+100, X+150,
X+200].
2. Compute your personalized tax rate as ((X % 5) + 5) / 100.
 Use broadcasting to apply this tax rate to each sales value.
3.  Adjust sales based on discount:
* If sales < X+100, apply a 5% discount.
* If sales >= X+100, apply a 10% discount.
4. Expand sales data for multiple weeks:
 * Create a 3×5 matrix representing three weeks of sales by stacking
sales three times using broadcasting.
 * Increase sales by 2% per week using element-wise broadcasting.

```python
import numpy as np


first_init = 'A'
last_init = 'B'

X = ord(first_init) + ord(last_init)


sales = np.array([X, X+50, X+100, X+150, X+200])


tax_rate = ((X % 5) + 5) / 100

sales_after_tax = sales * (1 - tax_rate)


discounts = np.where(sales < (X+100), 0.05, 0.10)
sales_after_discount = sales_after_tax * (1 - discounts)

weekly_sales = np.tile(sales_after_discount, (3, 1))

weeks = np.arange(3).reshape(3, 1)
weekly_sales *= (1 + 0.02 * weeks)

print("Original Sales:", sales)
print("Sales After Tax:", sales_after_tax)
print("Sales After Discount:", sales_after_discount)
print("Weekly Sales Matrix (After Increase):\n", weekly_sales)
```

Q4. Generate x values using np.linspace() from -10 to 10 with 100 points. Use
each function from the list below and compute y values using NumPy:
* Y = x2
* Y = sin(x)
* Y = ex
* Y = log(|x| + 1)

Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-10, 10, 100)

y1 = x ** 2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)


plt.figure(figsize=(10, 6))


plt.plot(x, y1, label="Y = x²", linestyle='dashed', color='r')
plt.plot(x, y2, label="Y = sin(x)", linestyle='solid', color='r')
plt.plot(x, y3, label="Y = e^x", linestyle='dotted', color='g')
plt.plot(x, y4, label="Y = log(|x| + 1)", linestyle='dashdot', color='purple')


plt.title("Mathematical Functions Visualization")
plt.xlabel("x values")
plt.ylabel("y values")

plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()

```
